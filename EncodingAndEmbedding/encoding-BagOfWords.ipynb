# Bag of Words (BoW)

**Bag of Words** is a text representation technique that converts documents into fixed-length numeric vectors by counting how many times each word from the vocabulary appears in a document.

### Key Idea
- Build a **vocabulary** of all unique words across all documents.
- For each document, create a vector where each element is the **count** of the corresponding word.
- The "bag" metaphor means word **order is ignored** — only frequency matters.

### Example
| Document | cricket | like | movie | people | watch |
|----------|---------|------|-------|--------|-------|
| people watch movie | 0 | 0 | 1 | 1 | 1 |
| people watch cricket | 1 | 0 | 0 | 1 | 1 |

### Why use it?
- Simple and fast to compute.
- Captures **word frequency** information unlike One Hot Encoding.

### Drawbacks
- **Ignores word order** and context — "not good" and "good not" are the same.
- **High dimensionality** for large vocabularies.
- **Common words** like 'and', 'the' dominate counts and carry little meaning.
- **No semantic understanding** — similar words have no relationship in vector space.

## Step 1: Import Libraries

`CountVectorizer` from **scikit-learn** implements the Bag of Words model. It handles tokenization, vocabulary building, and counting all in one step.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

## Step 2: Define the Corpus

A **corpus** is a collection of text documents. Notice that some words like `'watch'` and `'like'` appear **multiple times** within a single sentence — BoW will capture these repeat counts, unlike One Hot Encoding.

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

## Step 3: Create the CountVectorizer

`CountVectorizer()` initializes the BoW model with default settings:
- Converts text to **lowercase** automatically.
- Tokenizes by **whitespace and punctuation**.
- Ignores single-character tokens by default.

You can customize it with parameters like `stop_words='english'`, `max_features`, `ngram_range`, etc.

```python
vectorizer = CountVectorizer()
```

## Step 4: Fit and Transform the Documents

`.fit_transform()` does two things in one call:
1. **Fit** — scans all documents to build the vocabulary (unique word → column index mapping).
2. **Transform** — converts each document into a word-count vector.

The result is a **sparse matrix** (stored efficiently since most values are 0).

```python
bow_matrix = vectorizer.fit_transform(documents)
```

## Step 5: Inspect the Vocabulary

`.get_feature_names_out()` returns all unique words (features) learned during `.fit()`, sorted alphabetically.

Each word corresponds to one **column** in the BoW matrix.

```python
print("Vocabulary:", vectorizer.get_feature_names_out())
```

## Step 6: Display the BoW Matrix

`.toarray()` converts the sparse matrix to a dense **NumPy array** for readability.

Each **row** = one document.  
Each **column** = one word from the vocabulary.  
Each **cell value** = how many times that word appears in that document.

For example, `'watch'` appears twice in `"people watch movie and watch movie again"`, so its cell value is `2`.

```python
print("BOW Matrix:\n", bow_matrix.toarray())
```

## Summary

| Step | Action | Purpose |
|------|--------|---------|
| 1 | Import `CountVectorizer` | Load the BoW implementation from scikit-learn |
| 2 | Define corpus | Collect text documents with repeated words |
| 3 | Create vectorizer | Initialize `CountVectorizer()` |
| 4 | Fit and transform | Learn vocabulary & count word occurrences per document |
| 5 | Inspect vocabulary | See all unique words and their column positions |
| 6 | Display BoW matrix | View the word-count representation of each document |

> **BoW vs One Hot Encoding:** BoW captures **word frequency** (counts), while OHE only marks **presence** (0 or 1). For repeated words, BoW is more informative.
