# TF-IDF (Term Frequency – Inverse Document Frequency)

**TF-IDF** is a numerical statistic that reflects how important a word is to a document **relative to the entire corpus**. It improves on Bag of Words by penalizing common words that appear in many documents.

### The Formula

$$\text{TF-IDF}(t, d) = TF(t, d) \times IDF(t)$$

**Term Frequency (TF):** How often a term appears in a document.
$$TF(t, d) = \frac{\text{count of } t \text{ in } d}{\text{total words in } d}$$

**Inverse Document Frequency (IDF):** How rare a term is across all documents. Rare words get higher scores.
$$IDF(t) = \log\left(\frac{N}{1 + df(t)}\right) + 1$$

Where $N$ = total number of documents, $df(t)$ = number of documents containing term $t$.

### Key Idea
- Words that appear **frequently in one document** but **rarely across the corpus** get a **high** TF-IDF score → they are distinctive.
- Words that appear in **every document** (like `'and'`, `'the'`) get a **low** TF-IDF score → they carry little information.

### Why use it?
- Automatically **down-weights common stop words** without explicitly listing them.
- Captures **relative importance** of words, not just raw counts like BoW.
- Widely used in **information retrieval**, **search engines**, and **text classification**.

### Drawbacks
- Still **ignores word order** and context.
- Does not capture **semantic similarity** — synonyms are treated as different words.
- High dimensionality for large vocabularies.

## Step 1: Import Libraries

`TfidfVectorizer` from **scikit-learn** combines TF and IDF computation into a single, easy-to-use class. It handles tokenization, vocabulary building, TF-IDF scoring, and L2 normalization automatically.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

## Step 2: Define the Corpus

We use the same corpus as in the BoW notebook so results can be compared directly.

- Words like `'people'` appear in **all 4 documents** → low TF-IDF (common, not distinctive).
- Words like `'again'` or `'lot'` appear in only **1 document** → high TF-IDF (rare, distinctive).

```python
documents = ["people watch movie and watch movie again",
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

## Step 3: Create the TfidfVectorizer

`TfidfVectorizer()` initializes the vectorizer with default settings:
- Converts text to **lowercase**.
- Tokenizes by whitespace/punctuation.
- Applies **L2 normalization** to each row vector (so all row vectors have unit length).

You can customize with parameters like `stop_words='english'`, `max_features`, `ngram_range`, `smooth_idf`, etc.

```python
tfidf_vectorizer = TfidfVectorizer()
```

## Step 4: Fit and Transform the Documents

`.fit_transform()` does two things in one call:
1. **Fit** — builds the vocabulary and computes the IDF score for each word across all documents.
2. **Transform** — computes the TF-IDF score for every word in every document and returns a **sparse matrix**.

The final scores are also **L2-normalized** so that document length does not affect comparisons.

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)
```

## Step 5: Inspect the Vocabulary

`.get_feature_names_out()` returns all unique words learned during `.fit()`, sorted alphabetically.

Each word corresponds to one **column** in the TF-IDF matrix. The column order matches this vocabulary list.

```python
print("Vocabulary:", tfidf_vectorizer.get_feature_names_out())
```

## Step 6: Display the TF-IDF Matrix

`.toarray()` converts the sparse matrix to a dense **NumPy array** for readability.

Each **row** = one document.  
Each **column** = one word from the vocabulary.  
Each **cell value** = the TF-IDF score (a float between 0 and 1 after normalization).

**How to read the output:**
- A score of `0.0` means the word does **not appear** in that document.
- A **higher score** means the word is more **distinctive** for that document.
- Words like `'people'` that appear in all documents will have a **lower score** than rare words like `'again'`.

```python
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```
